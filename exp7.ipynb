```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Iris .csv")
```

```python
df.head()
```

```python
x=df.iloc[:, 1:5]
```

```python
x.head()
```

```python
from sklearn.cluster import  KMeans
```

```python
kmeans=KMeans(n_clusters=3).fit(x)
```

```python
import plotly.express as px
```

```python
px.scatter(x,color=kmeans.labels_)
```
